# COMPARISON USING AND / OR

```python
nameD = ['David Fincher']
nameF = ['Martin Scorsese']
Dave = netflix['director'].isin(nameD)
Martins = netflix['director'].isin(nameF)
rec = netflix['release_year'] > 2010
# afterwards we group them by functions >>> True or False
FN =netflix[(Dave | Martins) & rec]
# I sorted the release year 
FN.sort_values('release_year')
```

```python
# Another way to do this 
netflix[netflix['director'].isin(['David Fincher','Martin Scorsese'])].head()
```

# BITWISE NEGATION
## To negate the output 
### We use ~ (squiggle)


```python
df = titanic.head()
wom = df['sex'] == 'female'
df[~ wom]
```

```python
built = houses['yr_built'] >= 2014
renovated = houses['yr_renovated'] >= 2014
recent = built | renovated
houses[recent]
houses[~recent].head()
```

```python
nameD = ['David Fincher']
nameF = ['Martin Scorsese']
Dave = netflix['director'].isin(nameD)
Martins = netflix['director'].isin(nameF)
netflix[~(Dave | Martins)].head()
```

# BAR PLOTS

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
#SO IF YOU WANNA SPECIFY A PLOT WITH MATPLOT
#YOU CALL IT STRAIGHT

plt.figure(figsize=(6,6))
plt.bar(plants, germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
plt.style.use('seaborn-poster')
```

```python
#YOU CAN PUSH IT TO THE EDGE OF THE GRAPH 
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
#YOU CAN AND IT TO THE OTHER ONE 
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated ,label="Failed Seedlings")
plt.bar(plants, germinated, color="#2ecc71", label="Germinated Seedlings")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title("Greenhouse Seedlings")
plt.show()
```
